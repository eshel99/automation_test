```python
import mysql.connector
import pandas as pd

# Connection parameters-Bahamas
host = '107.23.185.153'
port = 3306
user = 'dba'
password = 'Hj47W7uYsK9!by^pE'
database = 'algo'


def update_stores_distributiongroup(connection, distributiongroup, storeno):
    try:
        cursor = connection.cursor()
        update_distributiongroup_query = f"UPDATE stores SET distributiongroup = {distributiongroup} WHERE StoreNo = '{storeno}';"
        cursor.execute(update_distributiongroup_query)
        connection.commit()
        cursor.close
    except mysql.connector.Error as e:
        print(f"Error updating distribution group for store {storeno}: {e}")

    


# Get user inputs:
distributiongroup = int(input("Enter the distributiongroup you want to commit: "))
stores = input("Enter the storeno, separate by comma: ").split(',')

# Establish a connection to the MySQL database
try:
    connection = mysql.connector.connect(
        host=host,
        port=port,
        user=user,
        password=password,
        database=database
    )

    if connection.is_connected():
        print(f"Connected to MySQL database '{database}'")

        # Loop through the stores and update the distributiongroup
        for storeno in stores:
            storeno = storeno.strip()   # Remove leading/trailing whitespaces if any
            # Call the function with the user-provided distributiongroup value
            update_stores_distributiongroup(connection, distributiongroup, storeno)
            connection.commit() 
            
        query = "SELECT storeno, distributiongroup FROM stores"
        # Execute a SELECT query to fetch the updated data
        updated_data = pd.read_sql_query(query, connection)
        print(updated_data)

except mysql.connector.Error as e:
    print(f"Error: {e}")

finally:
    # Closing the connection
    if 'connection' in locals() and connection.is_connected():
        connection.close()
        print("Connection closed")
```
